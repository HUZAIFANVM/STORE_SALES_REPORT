```python
df=df.drop(['Row ID','Order ID','Customer ID','Postal Code'],axis=1)
```

```python
df_sorted=df.groupby(['Category','Segment'])['Profit'].sum().sort_values(ascending=False)
```

```python
df_sort=df_sorted.to_frame()
```

```python
sns.set(style='whitegrid')
# Define the color codes
colors = ["#E8AFF6", "#AFB0F6", "#5566B8"]

# Set the Seaborn color palette
sns.set_palette(sns.color_palette(colors))
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Profit',hue='Segment',data=df_sort)
plt.xticks(rotation=45,ha='right')
plt.xlabel("Category",fontsize=12)
plt.ylabel("Profit",fontsize=12)
plt.title("Profit by Category And Segment",fontsize=14)
plt.tight_layout()
plt.savefig('/storage/emulated/0/pydriod3/Profit by Category.jpg',dpi=400)
plt.show()
```

```python
df_region=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
df_reg=df_region.to_frame()
```

```python
sns.set(style="whitegrid")
# Define the color codes
colors = ["#E8AFF6", "#AFB0F6", "#5566B8"]

# Set the Seaborn color palette
sns.set_palette(sns.color_palette(colors))
plt.figure(figsize=(10,6))
sns.barplot(x='Sales',y='Region',data=df_reg)
plt.title("Total Sales By Region")
plt.tight_layout()
plt.savefig('/storage/emulated/0/pydriod3/Total Sales By Region.jpg',dpi=400)
plt.show()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],format='mixed')
```

```python
df.info()
```

```python
df['Month']=df['Order Date'].dt.strftime('%B')
```

```python
df['Year']=df['Order Date'].dt.year
```

```python
df['Month']
```

```python
df_month=df.groupby('Month')['Sales'].sum().sort_values(ascending=False)
```

```python
df_mon=df_month.to_frame()
df_mon
```

```python
sns.set(style="whitegrid")
# Define the color codes
colors = ["#E8AFF6", "#AFB0F6", "#5566B8"]

# Set the Seaborn color palette
sns.set_palette(sns.color_palette(colors))
sns.barplot(x='Month',y='Sales',data=df_mon)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Total Sales Per Month")
plt.xticks(rotation=65)
plt.tight_layout()
plt.savefig('/storage/emulated/0/pydriod3/Total Sales Per Month.jpg',dpi=400)
plt.show()
```

```python
df['Months']=df['Order Date'].dt.month
```

```python
df_monthno=df.groupby('Months')['Sales'].sum().sort_values(ascending=False)
```

```python
df_monno=df_monthno.to_frame()
```

```python
sns.set(style="whitegrid")
# Define the color codes
colors = ["#E8AFF6", "#AFB0F6", "#5566B8"]

# Set the Seaborn color palette
sns.set_palette(sns.color_palette(colors))
plt.title("Monthly Sales Trend",fontsize=16)
sns.lineplot(x='Months',y='Sales',data=df_monno)
plt.xticks(rotation=65)
plt.tight_layout()
plt.savefig('/storage/emulated/0/pydriod3/Monthly Sales Trend.jpg',dpi=400)
plt.show()
```
